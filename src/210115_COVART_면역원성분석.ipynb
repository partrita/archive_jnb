```python
import pandas as pd

df = pd.read_csv('.\INPUT\\0115_COVART_prediction.csv', index_col=1)
df.tail()
```

# 데이터나누기

총 5가지 데이터로 나누기

```python
df_1 = df[df["ID"]=="ACE2_19-617_-GRAB3"]
df_2 = df[df["ID"]=="ACE2_19-106_-GRAB3"]
df_3 = df[df["ID"]=="ACE2_19-111_-GRAB3"]
df_4 = df[df["ID"]=="ACE2_19-111,385-395,321-362_-GRAB3"]
df_5 = df[df["ID"]=="CTC445.2-GRAB3"]
df_4.tail()
```

```python
# df_1 = df_1.drop(['ID','Pos'], axis=1).rename(index = lambda x : x[:1])
# df_1.tail()

def modify_data(df):
    temp_df = df.drop(['ID','Pos'], axis=1).rename(index = lambda x : x[:1])
    return temp_df

df_1 = modify_data(df_1)
df_2 = modify_data(df_2)
df_3 = modify_data(df_3)
df_4 = modify_data(df_4)
df_5 = modify_data(df_5)
```
