# Candidate 만들기

1. VH, VL 부분 뒤집어가면서
2. sc_Fv linker 2종 사용
3. PB101는 C-term 고정

````python
# tpa+A167_HC+linker_218+A167_LC+PB101
````

````python
seq = { 'A167': {'H': A167_HC, 'L': A167_LC},
        'A236': {'H': A236_HC, 'L': A236_LC},
        'KL00113':{'H': KL00113_VH,'L':KL00113_VL }
      }

linker1 = {'218':linker_218 ,
          'g4s':linker_g4s }
linker2 = 'GGGGS'
linker3 = 'GGGGSGGGGSGGGGS'
# seq['A169']['H']
````

````python
for i in ['A167','A236', 'KL00113']:
    for k in ['218','g4s']:
        Linker = linker1[k]
        for j in ['H','L']:
            if j == 'H':
                A = seq[i]['H']
                B = seq[i]['L']
                sequence = tpa+A+Linker+B+linker2+PB101
                print(f'>{i}_{j}_{k}_L')
                print(f'{sequence}')
            else:
                A = seq[i]['L']
                B = seq[i]['H']
                sequence = tpa+A+Linker+B+linker2+PB101
                print(f'>{i}_{j}_{k}_H')
                print(f'{sequence}')
````
