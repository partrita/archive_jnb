```python
import pandas as pd

df = pd.read_csv('./INPUT/210616_PB20X_kulen.csv', index_col=0)
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
import seaborn as sns
# Make the PairGrid
g = sns.PairGrid(df.sort_values("Titer (g/L)", ascending=False),
                 x_vars=df.columns[1:], y_vars=["Name"],
                 height=5, aspect=.3)

# Draw a dot plot using the stripplot function
g.map(sns.stripplot, size=10, orient="h", jitter=False,
      linewidth=1, edgecolor="w")

# Use the same x axis limits on all columns and add better labels
# g.set(xlim=(0, 25), xlabel="Yield", ylabel="")

# Use semantically meaningful titles for the columns
titles = ["HCCF", "ProteinA yield", "SEC yield",
          "Amount(mg)", "Monomer(%)"]

for ax, title in zip(g.axes.flat, titles):

    # Set a different title for each axes
    ax.set(title=title)

    # Make the grid horizontal instead of vertical
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

sns.despine(left=True, bottom=True)
```
