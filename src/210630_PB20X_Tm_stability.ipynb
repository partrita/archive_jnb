```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
import pandas as pd

df = pd.read_csv('./INPUT/210630_PB20X_Tm.csv')
df.head()
```

```python
plt.errorbar(df['Sample'],df['Tm_Mean'],yerr=df['Tm_Std'], fmt='o', capsize=0.2)
```

```python
import seaborn as sns

sns.set(style='ticks', context='talk')
sns.stripplot(None,'Tm_Mean', data=df, jitter=0.2)
sns.despine()
```

```python
sns.set(style='ticks')
plt.figure(figsize=(6, 6)) 
sns.stripplot(None, 'Tm_value', data=df, jitter=0.06)
plt.ylim(60, 70)
sns.despine()
```

```python
sns.boxplot(None, y="Tm_Mean", data=df)
```
