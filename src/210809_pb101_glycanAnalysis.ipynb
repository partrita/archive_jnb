```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import pandas as pd

df = pd.read_csv('./INPUT/210809_Panolos_CK1-5.csv')
df.tail()
```

```python
o_df = df[df['Types']=='Olinked']
n_df = df[df['Types']=='Nlinked']
n_df
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
# sns.set_theme(style="darkgrid")
# sns.despine(f, left=True, bottom=True)
# clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="m/z", y="Glyco_Sites",
                hue="Samples",
                palette="pastel",
                data=o_df, ax=ax)
```

```python
f, ax = plt.subplots(figsize=(16.5, 6.5))
# sns.set_theme(style="darkgrid")
# sns.despine(f, left=True, bottom=True)
# clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="m/z", y="Glyco_Sites",
                hue="Samples",
                palette="pastel", alpha=0.5,
                data=n_df, ax=ax)
```

```python
sns.relplot(
    data=df, x="m/z", y="Glyco_Sites",
    col="Types", hue="Samples", style="Samples",palette="pastel",
    kind="scatter"
)
```
