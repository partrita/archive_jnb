```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import pandas as pd

df = pd.read_csv('./INPUT/210827_PKB.csv')
df.tail()
```

```python
df[df["KD(M)"]]
```

```python
df['KD(nM)'] = df["KD(M)"]*1000000000
```

```python
# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df, kind="bar",
    x="Code_name", y="KD(nM)", hue="Code_name", col="Ligand",
    ci="sd", palette="dark", alpha=.6,
    sharey=False
)
# plt.xticks(rotation=70)
# g.despine(left=True)
# g.set_axis_labels("", "Body mass (g)")
# g.legend.set_title("")
g.set_xticklabels(rotation=90)
```

```python
# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df, kind="point",
    x="scFv", y="KD(nM)", hue="Code_name", col="Ligand",
    ci="sd", palette="Spectral", aplha = .5,
    sharey=False
)
# g.despine(left=True)
# g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("")
```

```python
# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df, kind="strip",
    x="scFv", y="KD(nM)", col="Ligand",
    ci="sd", palette="dark", alpha=0.6,
    sharey=False
)
# g.despine(left=True)
# g.set_axis_labels("", "Body mass (g)")
```
