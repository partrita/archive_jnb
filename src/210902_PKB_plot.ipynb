```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import pandas as pd

df = pd.read_csv('./INPUT/210907_PB20X_kulen.csv')
df.tail()
```

```python
sns.set_theme(style="whitegrid")
# Make the PairGrid
g = sns.PairGrid(df.sort_values("Titer (g/L)", ascending=False),
                 x_vars=df.columns[3:], y_vars=["Name"],
                 height=5, aspect=.35)

# Draw a dot plot using the stripplot function
g.map(sns.stripplot, size=7, orient="h", jitter=False,
      palette="hls", linewidth=1, edgecolor="w")

# Use the same x axis limits on all columns and add better labels
g.set(
#     xlim=(0, 25),
      xlabel="", ylabel="")

# Use semantically meaningful titles for the columns
titles = ["Titer(g/L)", "Purification yield(%)", "Final amount(mg)",
          "Purity(%)", "Tm(celsius)","VEGF binding(log10)","PlGF binding(fold)",
          "PD1 binding(fold)","PD-L1 binding(fold)"]

for ax, title in zip(g.axes.flat, titles):

    # Set a different title for each axes
    ax.set(title=title)

    # Make the grid horizontal instead of vertical
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

g.axes[0,0].set_xlim(0,0.7)
g.axes[0,1].set_xlim(0,60)
# g.axes[0,2].set_xlim(0,5)
g.axes[0,3].set_xlim(0,105)
g.axes[0,4].set_xlim(60,70)
g.axes[0,5].set_xlim(-3,1)
g.axes[0,6].set_xlim(0,3)
g.axes[0,7].set_xlim(0,2.5)
g.axes[0,8].set_xlim(0,2)

sns.despine(left=True, bottom=True)
```
