```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import pandas as pd

df = pd.read_csv('./INPUT/210916_PB20X.csv')
df
```

```python
df_pb10x = df[df['category']=='PB10X']
df_pb20x = df[df['category']=='PB20X']
```

```python
plt.style.use(['ggplot'])
plt.figure(figsize=(10,6))
plt.ylim(0,14)
plt.bar(df_pb10x['name'],df_pb10x['Eluate amounts\n(mg)'])
```

```python
plt.style.use(['ggplot'])
plt.figure(figsize=(10,6))
plt.ylim(0,14)
plt.bar(df_pb20x['name'],df_pb20x['Eluate amounts\n(mg)'])
```

```python
plt.style.use(['ggplot'])
f, ax = plt.subplots(figsize=(6, 15))
crashes = df.sort_values("Monomer", ascending=False)

# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(y="name", x="Monomer", data=crashes,
            label="Monomer", color="b")

# Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(y="name", x="HMWS", data=crashes,
            label="HMWs", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(
#     xlim=(0, 24),
       ylabel="",
       xlabel="")
sns.despine(left=True, bottom=True)
```
