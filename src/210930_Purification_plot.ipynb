```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import pandas as pd

df = pd.read_csv('./INPUT/210930_PB101_SBL.csv')
df
```

```python
plt.style.use(['ggplot'])
# sns.set_theme(style="ticks")
sns.barplot(x='Step', y='Monomer', hue='Batch', data=df)
```

```python
# Set up the matplotlib figure
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 7), sharex=True)

# Generate some sequential data
# x = np.array(list("ABCDEFGHIJ"))
# y1 = np.arange(1, 11)

sns.barplot(x='Step', y='Monomer', hue='Batch', palette="rocket", data=df, ax=ax1)
# ax1.axhline(0, color="k", clip_on=False)
# ax1.set_ylabel("Sequential")
ax1.set_xlabel("")
ax1.get_legend().remove()
ax1.set_ylim([50,100])

# Center the data to make it diverging
sns.barplot(x='Step', y='HCP_ppm', hue='Batch', palette="rocket", data=df, ax=ax2)
# ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("HCP(ppm)")
ax2.get_legend().remove()
ax2.set_ylim([0, 1000])

# Randomly reorder the data to make it qualitative
sns.barplot(x='Step', y='HCP_ppb', hue='Batch', palette="rocket",data=df,  ax=ax3)
# ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("HCD(ppb)")
ax3.set_ylim([0, 500])

# Finalize the plot
# sns.despine(bottom=True)
# plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=2)
```

```python
# Set up the matplotlib figure
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 7), sharex=True)

# Generate some sequential data
# x = np.array(list("ABCDEFGHIJ"))
# y1 = np.arange(1, 11)

sns.barplot(x='Step', y='Monomer', hue='Batch', data=df, ax=ax1)
# ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Purity(%)")
ax1.set_xlabel("")
ax1.get_legend().remove()
ax1.set_ylim([70,100])

# Center the data to make it diverging
sns.barplot(x='Step', y='HCP_ppm', hue='Batch', data=df, ax=ax2)
# ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("HCP(ppm)")
ax2.set_xlabel("")
ax2.get_legend().remove()
ax2.set_yscale('log', base=10)
# ax2.set_ylim([0, 100])

# Randomly reorder the data to make it qualitative
sns.barplot(x='Step', y='HCP_ppb', hue='Batch', data=df,  ax=ax3)
# ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("HCD(ppb)")
ax3.set_yscale('log', base=10)
# ax3.set_ylim([0, 3])

# Finalize the plot
# sns.despine(bottom=True)
# plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=2)
```
