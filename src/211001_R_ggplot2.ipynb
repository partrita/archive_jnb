```r
library(ggplot2)
library(repr)
```

```r
g <- ggplot(
    data = iris,
    mapping = aes(x = Sepal.Length, y = Sepal.Width)) + geom_point()

g + annotate(geom = "text", x = iris$Sepal.Length, y = iris$Sepal.Width, label = rownames(iris), colour = "brown", alpha = 0.7, size = 3, hjust = 0.5, vjust = -1)
options(repr.P.width=1,repr.P.height=1)
```

```r
fig <- function(width, heigth){
     options(repr.plot.width = width, repr.plot.height = heigth)
}
```

```r
df_1 = data.frame(x=c(5, 6, 7, 8, 9), y = c(200, 225, 250, 270, 310))
options(repr.plot.width = 3, repr.plot.height = 3) 
ggplot(df_1, aes(x = x, y = y)) + geom_point(size = 1.3)
```
