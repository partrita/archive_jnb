# 가설 검정
가설 검정은 효과 검증의 기반으로, 표본을 통해 모집단에 유의한 차이가 있는지 확인하는 방법입니다.

## 동전 던지기

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats
import pandas as pd
```

50% 확률로 앞면이 나오는 동전을 42번 던져본 결과의 분포를 시각화 해보면 다음과 같습니다.

```python
x = np.arange(1,11)
y = scipy.stats.binom.pmf(x, 10, 0.5)
plt.bar(x,y)
plt.xlabel('coin head')
plt.ylabel('')
```

앞면이 5번 이 나올 확률이 가장 높음을 알 수 있다.

# 가설 검정에서 주의할 점

가설 검정을 잘못 사용하면 거짓 양성의 확률이 높아져 p-value만으로 결정을 내리면 본질을 놓칠 수 있습니다.

## 반복 검정
가설 검정에서 주의할 점은 검정 대상 표본을 고정해야 한다는 것입니다. 유의미한 차이를 찾기 위해 표본을 바꿔가면 반복하면 결국에는 p-value를 낮출 수 있기 때문입니다.

예시로 동전 던지기의 예를 살펴보겠습니다.

```python
mu = 0.5 # 확률 50%
init_sample = list(scipy.stats.bernoulli.rvs(mu, size = 10))

sample = init_sample
p_value_history = []
for i in range(100):
    _, p_value = scipy.stats.ttest_1samp(sample[-20:], 0.5)
    p_value_history.append(p_value)
    sample.append(scipy.stats.bernoulli.rvs(mu))

plt.figure(figsize = (10,5))
plt.plot(p_value_history)
plt.xlabel('test epoch')
plt.ylabel('p_value')
```

그림을 보면 p-값이 0.05 아래로 내려갔습니다. 앞면이 나올 확률은 50%으로 고정되어 있는데 검정을 반복하다 보면 언젠가는 유의한 차이가 나오게 됩니다.

## 표본의 크기

표본의 크기를 계속 늘려가면 작은 차이로도 유의한 차이를 만들 수 있습니다. 아래의 예시를 확인해 보시죠.

```python
max_sample = 1000
# 표본 a의 평균을 50%, b의 평균을 50.1%로 설정합니다.
a = scipy.stats.bernoulli.rvs(0.500, size = max_sample)
b = scipy.stats.bernoulli.rvs(0.501, size = max_sample)
p_values = []
```

```python
# 표본의 크기를 10000씩 늘려가며 검정
sample_sizes = np.arange(0, max_sample, 100)
for sample_size in sample_sizes:
    _, p_value = scipy.stats.ttest_ind(a[:sample_size], b[:sample_size], equal_var = False)
    p_values.append(p_values)

    
plt.figure(figsize = (10,4))
plt.plot(sample_sizes, p_values)
plt.xlabel('size of sample')
plt.ylabel('p_value')
```
