```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
df.tail()
```

```python
df.describe().T
```

```python
plt.scatter(df['alcohol'], df['quality'])
plt.ylabel('quality')
plt.xlabel('alcohol')
```

```python
plt.scatter(df['volatile acidity'], df['quality'])
plt.ylabel('quality')
plt.xlabel('volatile acidity')
```

```python
import seaborn as sns
# plt.figure(figsize=(15,4))
ax  = sns.pairplot(df, markers="+",
             y_vars=['fixed acidity','volatile acidity', 'citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol'], 
             x_vars=['quality'], aspect=2)
```

# 모델 만들기


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# y = df['quality']
y = df.pop('quality')
# X = df[list(df.columns)[:-1]]
X = df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
reg_clf = LinearRegression()
reg_clf.fit(X_train, y_train)
y_pred = reg_clf.predict(X_test)
print('R-squared: {}'.format(reg_clf.score(X_test,y_test)))
```

```python
plt.scatter(y_test, y_pred)
plt.ylim(3,8)
plt.ylabel('predicted quality')
plt.xlabel('true quality')
```
