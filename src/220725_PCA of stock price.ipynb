PCA는 다양한 분야에서 사용된다. 이번에는 금유분야에서 어떻게 쓰이는지 알아보자. 우선 미국(US), 일본(JP), 유럽(EZ), 한국(KR)의 과거 20년간 주가를 살펴보자

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.decomposition import PCA
import pandas as pd
# pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
# from pandas_datareader import data, wb
import datetime

symbols = [
    "SPASTT01USM661N", 
    "SPASTT01JPM661N", 
    "SPASTT01KRM661N",
]
data = pd.DataFrame()
for sym in symbols:
    data[sym] = web.DataReader(sym, data_source='fred', 
                               start=datetime.datetime(1981, 1, 1),
                               end=datetime.datetime(2018, 8, 31))[sym]
data.columns = ["US", "JP", "KR"]
data = data / data.iloc[0] * 100
```

```python
import matplotlib.pyplot as plt
data.plot()
plt.title("Stock price of country")
plt.show()
```

앞의 차트는 시작시점의 주가가 100이 되도록 크기를 조정하였다. 이 데이터로부터 연간 주식수익률 데이터를 구하면 다음처럼  20×4  크기의 특징행렬을 구할 수 있다.

```python
data
```

```python
df = ((data.pct_change() + 1).resample('AS').prod() - 1).T * 100
# df = data.resample('AS').mean()
df.iloc[:, :5]
```

이 연간 주식수익률 데이터를 나타내면 다음과 같다. 붓꽃의 경우보다 (2차원에서 20차원으로) 차원은 증가했지만 각 표본이 가지는 측정 데이터 벡터의 모양(비율)이 비슷하다는 것을 알 수 있다. 붓꽃이 측정 데이터의 값을 결정하는 "크기"라는 잠재변수를 가지고 있는 것처럼 각 나라의 수익률은 "경제적 요인"이라고 하는 공통된 잠재변수에 의해서 결정되기 때문이다.

```python
df.T.plot()
plt.title("Earning rate of year")
plt.xticks(df.columns)
plt.show()
```

PCA분석을 통해 모든 나라의 주가 움직임 기저에 깔려있는 경기의 변동을 살펴보자.
일단 세계주가의 공통요인은 평균값으로 구할 수 있다.

```python
pca2 = PCA(n_components=1)
w = pca2.fit_transform(df)
```

```python
m = pca2.mean_
```

그리고 나라별로 주가를 다르게 하는 요인은 주성분으로 구할 수 있다.

```python
p1 = pca2.components_[0]
```

이 두가지 성분을 차트로 나타내보자. 평균주가로부터 주성분이 더해질수록 어떻게 주가의 형태가 바뀌는지 살펴보면 다음과 같다. 굵은 선으로 표시된 주가가 평균 주가이고 나머지 주가는 주성분이 점점 많이 추가됨에 따라 어떤 주가 형태로 변하는지를 표시한 것이다.

```python
import numpy as np
for i in np.linspace(0, 100, 10):
    plt.plot(range(1981, 2019), pca2.mean_ + p1 * i) # range 를 맞추기 위해 2019 사용
plt.plot(range(1981, 2019), pca2.mean_, "ko-", lw=3)
# plt.title("주성분 크기에 의한 주가의 변화")
plt.show()
```

이 주성분이 가지는 의미를 살펴보기 위해 각 나라의 주성분 비중과 주성분을 이용하여 구한 근사 주가를 표시하면 다음과 같다.

```python
df_w = pd.DataFrame(w)
df_w.index = df.index
df_w.columns = ["Principal component ratio"] # 주성분비중
df_w
```

```python
df_i = pd.DataFrame(pca2.inverse_transform(w))
df_i.index = df.index
df_i.columns = df.columns
df_i.iloc[:, -10:]
df_i.T.plot()
# plt.title("주성분을 사용한 주요국의 과거 20년간 연간수익률 근사치")
plt.xticks(df.columns)
plt.show()
```

이 결과로부터 우리가 구한 주성분은 (미국, 일본) 이라는 데이터로부터 (한국)이라는 데이터로 변화시키는 요인 즉 중진국 요인이라 부를 수 있는 성분이라는 것을 알 수 있다.
