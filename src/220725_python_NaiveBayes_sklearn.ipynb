# sklearn으로 구현한 나이브 베이즈

# 데이터셋 설명

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)
```

```python
lr_clf = LogisticRegression()
nb_clf = GaussianNB()

score_lr , score_nb = ([] for i in range(2))

train_sizes = range(10, len(X_train), 25)

for train_size in train_sizes:
    X_slice, _, y_slice, _ = train_test_split(X_train, y_train, train_size=train_size,
                                             stratify=y_train, random_state=42)
    nb_clf.fit(X_slice,y_slice)
    score_nb.append(nb_clf.score(X_test, y_test))
    lr_clf.fit(X_slice, y_slice)
    score_lr.append(lr_clf.score(X_test, y_test))
```

```python
plt.plot(train_sizes, score_nb, label='Naive Bayes')
plt.plot(train_sizes, score_lr, label='Logistic Regression')
plt.xlabel('Number of training instances')
plt.ylabel('Test set accuracy')
plt.legend()
```

데이터셋이 작을 수록 대개 나이브 베이즈가 로지스틱 회귀보다 더 뛰어난 성을 보여줍니다. 이 예제에서는 처음에는 나이브 베이즈가 로지스틱 분류기보다 더 뛰어난 성능을 보이지만 훈련 데이터 크기가 커지면 로지스틱 회귀의 성능이 많이 개선되는 것을 볼 수 있습니다.

# 피마 인디언 당뇨 데이터

```python
df = pd.read_csv('./data/diabetes.csv')
df.tail()
```

```python
y = df['Outcome']
X = df.iloc[:, :8]
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

lr_clf = LogisticRegression()
nb_clf = GaussianNB()

score_lr , score_nb = ([] for i in range(2))

train_sizes = range(10, len(X_train), 10)

for train_size in train_sizes:
    X_slice, _, y_slice, _ = train_test_split(X_train, y_train, train_size=train_size,
                                             stratify=y_train, random_state=42)
    nb_clf.fit(X_slice,y_slice)
    score_nb.append(nb_clf.score(X_test, y_test))
    lr_clf.fit(X_slice, y_slice)
    score_lr.append(lr_clf.score(X_test, y_test))

plt.plot(train_sizes, score_nb, label='Naive Bayes')
plt.plot(train_sizes, score_lr, label='Logistic Regression')
plt.xlabel('Number of training instances')
plt.ylabel('Test set accuracy')
plt.legend()
```

훈련 데이터의 크기를 증가시키면서 두 모델의 정확도 그림을 그려보면, 역시 데이터의 크기가 작을때 나이브베이즈의 성능이 좋고 데이터가 커질수록 로지스틱 분류기의 성능이 지속적으로 개션되는 것을 볼 수 있습니다.
