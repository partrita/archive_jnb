To do linear regression on pandas columns: from scipy.stats import linregress slope, intercept, rvalue, pvalue, stderr = linregress(df['foo'], df['bar']) Useful when combined with seaborn's regplot()

```python
from scipy.stats import linregress
import seaborn as sns
import pandas as pd 

df = pd.DataFrame({'x': [3.52,2.58,3.31,4.07,4.63,3.98,4.29,4.83,3.71,4.61,3.90,3.20],
'y': [2.48,2.27,2.47,2.77,2.98,3.05,3.18,3.46,3.03,3.25,2.67,2.53]})

slope, intercept, rvalue, pvalue, stderr = linregress(df['x'], df['y'])

p1 = sns.regplot(x = df['x'], y = df['y'])
# add text annotation
p1.text(2.5, 3.5, 'P-value: {:.4f}'.format(pvalue), horizontalalignment='left', size='medium', color='black')
```
