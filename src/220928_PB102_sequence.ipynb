```python
# Importing the matplotlib.pyplot
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
import pandas as pd
d = {'primer':['AA1_F1','AA1_F2','AA1_F3','AA1_R1'] , 'start': [91, 283,454,1], 'end': [444, 471,471,124]}
df = pd.DataFrame(data=d)
df['left'] = df['end'] - df['start']
df
```

```python
fig, ax = plt.subplots(1, figsize=(16,6))
plt.xlim(0,471)
ax.barh(df.primer, df.left, left=df.start)
plt.show()
```
